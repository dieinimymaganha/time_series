```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

Carregando a base de co2

```python
co2 = pd.read_csv('sample_data/co2.csv', delimiter='\t')
co2.head()
```

Carregando a base de Nascimento

```python
plt.plot(co2.data, co2.media)
plt.ylabel('MÃ©dia')
plt.xlabel('Data')
```

```python
nasc = pd.read_csv('sample_data/nascimentos.csv')
nasc.head()
```

```python
plt.plot(nasc.data, nasc.n_nasc)
plt.ylabel('Nascimento')
plt.xlabel('Data')
```

Convertendo coluna data para tipo datetime

```python
nasc.data = pd.to_datetime(nasc.data)
```

```python
plt.plot(nasc.data, nasc.n_nasc)
plt.ylabel('Nascimento')
plt.xlabel('Data')
```

# PrevisÃ£o atravÃ©s da mÃ©dia

### CO2

```python
media_carb = np.mean(co2.media)
media_carb
```

```python
dt_carb = co2.data[1] - co2.data[0]
```

```python
plt.plot(co2.data, co2.media)
plt.plot(co2.iloc[-1,0] + dt_carb, media_carb,'*')
plt.ylabel('MÃ©dia')
plt.xlabel('Data')
```

```python
#### mÃ©dia dos ultimos 3 pontos
```

```python
media_tres_pontos_c = np.mean(co2.media[474:476])
plt.plot(co2.data, co2.media)
plt.plot(co2.iloc[-1,0] + dt_carb, media_tres_pontos_c,'*')
plt.ylabel('MÃ©dia')
plt.xlabel('Data')
```

### Nascimento

```python
media_nasc = np.mean(nasc.n_nasc)
dt_nasc = nasc.data[1] - nasc.data[0]
```

```python
plt.plot(nasc.data, nasc.n_nasc)
plt.plot(nasc.iloc[-1,0] + dt_nasc, media_nasc,'*')
plt.ylabel('Nascimento')
plt.xlabel('Data')
```

```python
media_tres_pontos_nasc = np.mean(nasc.n_nasc[362:364])
plt.plot(nasc.data, nasc.n_nasc)
plt.plot(nasc.iloc[-1,0] + dt_nasc, media_tres_pontos_nasc,'*')
plt.ylabel('Nascimento')
plt.xlabel('Data')
```
